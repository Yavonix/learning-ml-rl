```python
import numpy as np
from typing import Iterable, Callable
```

```python

v = np.zeros((4,4))
states = [(i, j) for i in range(len(v)) for j in range(len(v)) if (i, j) not in {(0, 0), (3, 3)}]
threshold = 1e-3
gamma = 1
```

```python
while True:

    delta = 0

    for (i, j) in states:
        v_old = v[(i,j)]
        v[(i,j)] = -1 + 0.25 * gamma * (v[(min(i+1,3),j)] + v[(max(i-1,0),j)] + v[(i,min(j+1, 3))] + v[(i,max(j-1, 0))])
        delta = max(delta, v_old - v[(i,j)])

    if delta < threshold:
        break
```
