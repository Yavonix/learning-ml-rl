### First, we import torch. Note that although it's called PyTorch, we should import torch instead of pytorch.

```python
import torch
from jax import numpy as jnp
```

### A tensor represents an array of values, which may have multiple dimensions.

```python
x = torch.arange(12)

xj = jnp.arange(12)

x, xj
```

### The shape of a tensor and the total number of elements in the tensor can be accessed:

```python
x.shape, xj.shape
```

```python
x.numel(), xj.size
```

### To change the shape of a tensor without changing the number of elements or their values, you can call the `reshape` function.

```python
x.reshape(3,4), xj.reshape((3,4))
```

### Create special tensors

```python
torch.zeros(2,3,4), jnp.zeros((2,3,4))
```

```python
torch.ones(2,3,4), jnp.ones((2,3,4))
```

### Randomly create tensors using the uniform distribution:

```python
from flax import nnx
import jax

key = jax.random.key(42)

key, norm_key = jax.random.split(key)


torch.rand(3,4), jax.random.uniform(key=norm_key, shape=(3,4), minval=0, maxval=1)
```

```python
torch.randn(3,4), Rngs.normal((3,4))
```

### Create a tensor from a given ndarray (in Python)

```python
torch.tensor([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]]), jnp.array([[2, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
```

### Standard arithmetic operators (+, -, *, /, and **) can all be listed to element-wise operations for tensors.

```python
x = torch.tensor([1.0, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])

xj = jnp.array([1.0, 2, 4, 8])
yj = jnp.array([2, 2, 2, 2])

x + y, x - y, x * y, x / y, x**y, torch.exp(x), xj + yj, xj - yj, xj * yj, xj / yj, xj**yj, jnp.exp(xj)
```

### Concatenating tensors

```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])
print(torch.cat((X, Y), dim=0), "\n", torch.cat((X, Y), dim=1))

XJ = jnp.array(X)
YJ = jnp.array(Y)

print(jnp.concat((XJ, YJ), axis=0), "\n", jnp.concat((XJ, YJ), axis=1))
```

### Creating binary tensors using logical operators

```python
X == Y, XJ == YJ
```

```python
X > Y, XJ > YJ
```

```python
X < Y, XJ < YJ
```

### Summing all elements of a tensor produces a tensor with only one element.

```python
X.sum(), XJ.sum()
```

### Broadcasting Mechanism: Performing tensor operations for incompatible sizes of tensors

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))

aj = jnp.array(a)
bj = jnp.array(b)

a, b
```

For different shapes, the shape values with 1 get copied.
For different sizes, we right align then add 1:
- adding (2, ) to (3, 1)
- Right align: (   2), (3, 1)
- Fill empty spots with 1: (1, 2), (3, 1)
- Applying copying rule

```python
a+b, aj + bj
```

### To assign the same value to multiple elements, we simply need to index all the elements and then assign the value to them.

```python
X[1:3, :] = 15
print(X)
XJ = XJ.at[1:3, :].set(15)
print(XJ)
```
