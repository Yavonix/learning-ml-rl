### Exercise 7: Stack and Slice Three Matrices

Let `A`, `B`, and `C` be matrices of shape `100 x 200`.

Tasks:
1. Stack them into one tensor with a new axis
2. Report the resulting tensor shape
3. Recover matrix `B` by slicing
4. Verify recovery (for example with `torch.equal`)

Hint: Decide which axis should represent "which matrix" before stacking, then slice along that same axis to retrieve `B`.

```python
# TODO: Stack A, B, C and recover B by slicing
A = torch.randn((100, 200))
B = torch.randn_like(A)
C = torch.randn_like(A)

D = torch.stack((A, B, C))

print(D.shape)

B_recov = D[1, :, :]

print(torch.equal(B, B_recov))
```
