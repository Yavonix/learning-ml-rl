```python
import torch
from jax import numpy as jnp
from flax import nnx
rngs = nnx.Rngs(42)
```

Pytorch errors when stride $\ne 1$ and padding = same

```python
conv_torch = torch.nn.Conv2d(3, 1, (3,3), padding='same', stride=2)
```

Flax nnx fine with stride $\ne 1$ and padding = same

```python

conv_flax = nnx.Conv(3, 1, (3,3), rngs=rngs, padding="SAME", strides=(2,2))
```

Also by default conv layers do have bias:
