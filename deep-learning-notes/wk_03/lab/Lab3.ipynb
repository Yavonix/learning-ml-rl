# Lab 3: Implement MLP and Conquer Overfitting

### Objective: 

In this assignment, you will build a Multilayer Perceptron (MLP) to classify a dataset. First, you will intentionally build a model with high capacity to see how it overfits the training data. Then, you will use the regularization techniques discussed in class—Weight Decay and Dropout —to "tame" your model and improve its generalization error.

### Part 1: Loading and Visualizing Data

### Part 1: Data Preparation

1. Load the Dataset: We will use the classic FashionMNIST dataset (available via torchvision.datasets).
2. Split the Data: Create a Training set, a Validation set (10% of the training data), and a Test set.

Make sure that your validation dataset is absolutely not mixed with your training data.

### Part 2: Build the Baseline Model (The Memorizer)

1. Architecture: Build an MLP with at least three hidden layers. Make the hidden layers wide (e.g., 512 neurons each) so the model has a high capacity.
2. Training: Train the model using standard Stochastic Gradient Descent (SGD) without any regularization.
3. Tracking: Plot both the Training Loss and Validation Loss over your epochs. You should clearly see the point where the training loss continues to drop while the validation loss plateaus or spikes (the classic signature of overfitting).


### Apply Regularization (The Fix)

Now, create a second model (Regularised_MLP) with the exact same architecture, but apply our two new techniques:

1. Dropout: Add a Dropout layer after your fully-connected hidden layers. Use a probability $p = 0.5$. Remember to use model.train() and model.eval() so dropout is only active during training and bypassed during inference.
2. Weight Decay: When setting up your SGD optimizer, add the weight_decay parameter (this represents our $\lambda$ penalty). Try a value like $1e-4$.
3. Training: Train this new model for the same number of epochs and plot the Training vs. Validation loss.
