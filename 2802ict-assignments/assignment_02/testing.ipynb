```python
import polars as pl
import math
from dataclasses import dataclass
from typing import Literal, Union, cast

Status = Literal["unacc", "acc", "good", "vgood"]
    
car_columns = {
    "buying": pl.Enum(["low", "med", "high", "vhigh"]),
    "maint": pl.Enum(["low", "med", "high", "vhigh"]),
    "doors": pl.Enum(["2", "3", "4", "5more"]),
    "persons": pl.Enum(["2", "3", "4", "more"]),
    "lug_boot": pl.Enum(["small", "med", "big"]),
    "safety": pl.Enum(["low", "med", "high"]),
    "class": pl.Enum(["unacc", "acc", "good", "vgood"]),
}
```

```python
examples = pl.read_csv("./dt-car.csv", schema_overrides=car_columns)
```

```python
examples["class"].n_unique()
```

```python
outcomes = examples["class"].unique()
total = examples.height

for i in outcomes:
    outcome_chance = examples.filter(pl.col("class") == i).height / total
    print(outcome_chance)
```

```python
list(map(lambda k: examples.filter(pl.col("class") == k).height / total, outcomes))
```
