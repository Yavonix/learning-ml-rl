## Jupyter notebook for testing/benchmark the NN

```python
from nn import Model, ModelMetrics, init_model, train, eval, sep_labels, batch
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from tqdm import tqdm # for progress bar
import itertools

n_input = 784
n_hidden = 30
n_output = 10
classes = 10

d_train = sep_labels(np.loadtxt("/home/roman/learning-ml/2802ict-assignments/assignment_02/nn/fashion-mnist_train.csv.gz", delimiter=',', skiprows=1), classes) # (60000, 785) first col is label
d_test = sep_labels(np.loadtxt("/home/roman/learning-ml/2802ict-assignments/assignment_02/nn/fashion-mnist_test.csv.gz", delimiter=',', skiprows=1), classes) # (10000, 785) first col is label
```

Part 1: epochs = 30, mini-batch size = 20, 𝜂 = 3; graph of test accuracy vs epoch; print max accuracy

```python
model = init_model(n_input, n_hidden, n_output)
metrics = ModelMetrics()

lr = 3
n_epochs = 30
batch_size = 20

data_1 = {'epoch': np.arange(n_epochs), 'val': [], 'train': []}

for i in tqdm(range(n_epochs)):
    # mb mini batch
    for mb in batch(d_train, batch_size):
        train(mb, model, metrics, lr)
    data_1['train'].append(metrics.accuracy.value)
    metrics.reset()

    for mb in batch(d_test, batch_size):
        eval(mb, model, metrics)
    data_1['val'].append(metrics.accuracy.value)
    metrics.reset()

print(f"max acc={max(data_1['val']):.2f}")
fig = px.line(data_1, x='epoch', y=['val', 'train'],
			  labels={'value': 'Accuracy', 'epoch': 'Epoch', 'variable':'Series'},
			  title=f"Test Accuracy vs Epoch (lr={lr}, batch_size={batch_size})")
fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
fig.show()
```

Part 2: epochs = 30, mini-batch size = 20, 𝜂 = 0.001, 0.01, 1.0, 10, 100; graph of test accuracy vs epoch for each 𝜂; print max accuracy for each 𝜂

```python
metrics = ModelMetrics()

learning_rates = [0.001, 0.01, 1.0, 10, 100]
n_epochs = 30
batch_size = 20

data_2 = {f'{k} lr={i}': [] for i in learning_rates for k in ["train", "val"]}
data_2['epoch'] = np.arange(n_epochs).tolist()

for lr, epoch in tqdm(list(itertools.product(learning_rates, range(n_epochs)))):
    if epoch == 0: model = init_model(n_input, n_hidden, n_output) # reset model
    # mb mini batch
    for mb in batch(d_train, batch_size):
        train(mb, model, metrics, lr)
    data_2[f'train lr={lr}'].append(metrics.accuracy.value)
    metrics.reset()

    for mb in batch(d_test, batch_size):
        eval(mb, model, metrics)
    data_2[f'val lr={lr}'].append(metrics.accuracy.value)
    metrics.reset()
```

```python
print("\n".join(f'lr={i} max acc={max(data_2[f'val lr={i}']):.2f}' for i in learning_rates))

fig = px.line(data_2, x='epoch', y=[f'val lr={i}' for i in learning_rates],
			  labels={'value': 'Accuracy', 'epoch': 'Epoch', 'variable':'Series'},
			  title=f"Test Accuracy vs Epoch (lr={learning_rates}, batch_size={batch_size})")
fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
fig.show()
```

Part 3: epochs = 30, mini-batch size = 1, 5, 20, 100, 300, 𝜂 = 3; graph of test accuracy vs epoch for each mini-batch size; print max accuracy for each mini-batch

```python
metrics = ModelMetrics()
model = init_model(n_input, n_hidden, n_output)

lr = 3
n_epochs = 30
batch_sizes = [1, 5, 20, 100, 300]

data_3 = {f'{k} mb={i}': [] for i in batch_sizes for k in ["train", "val"]}
data_3['epoch'] = np.arange(n_epochs).tolist()

for batch_size, epoch in tqdm(list(itertools.product(batch_sizes, range(n_epochs)))):
    if epoch == 0: model = init_model(n_input, n_hidden, n_output) # reset model
    # mb mini batch
    for mb in batch(d_train, batch_size):
        train(mb, model, metrics, lr)
    data_3[f'train mb={batch_size}'].append(metrics.accuracy.value)
    metrics.reset()

    for mb in batch(d_test, batch_size):
        eval(mb, model, metrics)
    data_3[f'val mb={batch_size}'].append(metrics.accuracy.value)
    metrics.reset()
```

```python
print("\n".join(f'mb={i} max acc={max(data_3[f'val mb={i}']):.2f}' for i in batch_sizes))

fig = px.line(data_3, x='epoch', y=[f'val mb={i}' for i in batch_sizes],
			  labels={'value': 'Accuracy', 'epoch': 'Epoch', 'variable':'Series'},
			  title=f"Test Accuracy vs Epoch (lr={lr}, batch_size={batch_sizes})")
fig.update_layout(title={'x': 0.5, 'xanchor': 'center'})
fig.show()
```
