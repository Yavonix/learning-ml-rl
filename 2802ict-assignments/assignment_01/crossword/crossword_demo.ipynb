## Crossword Demo

```python
from crossword import Crossword
from crossword_partly_filled import CrosswordPartlyFilled
from generate import CrosswordCreator
from pathlib import Path
import ipywidgets as widgets
from IPython.display import display

structure_selector = widgets.Dropdown(
    options=["structure0", "structure1", "structure2", "structure2_partly_filled"],
    value='structure0',
    description='Structure File:',
    disabled=False,
)
display(structure_selector)

word_list_selector = widgets.Dropdown(
    options=["words0", "words1", "words2"],
    value='words0',
    description='Word File:',
    disabled=False,
)
display(word_list_selector)
```

```python
structure = Path(f"./data/{structure_selector.value}.txt")
words = Path(f"./data/{word_list_selector.value}.txt")

crossword = Crossword(structure, words)
creator = CrosswordCreator(crossword)

assignment = creator.solve()
if assignment: creator.save(assignment)
else: print("No assignment")
```
