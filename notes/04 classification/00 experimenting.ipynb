```python
import jax
from jax import numpy as jnp
```

```python
def cross_entropy(y_hat: jax.Array, y: jax.Array):
    # - sum log(y_hat) * y
    rows = jnp.arange(y.shape[0])
    return -jnp.log(y_hat[rows, y]).mean()
```

```python
from flax import nnx
```

```python
metrics = nnx.MultiMetric(
    loss = nnx.metrics.Average("loss"),
    acc = nnx.metrics.Accuracy()
)

metrics.update(loss=0.5, logits=jnp.array([[0.1,0.9]]), labels=jnp.array([1]))
metrics.update(loss=0.1, logits=jnp.array([[0.1,0.9]]), labels=jnp.array([0]))
metrics.update(loss=0.1, logits=jnp.array([[0.1,0.9]]), labels=jnp.array([0]))

m = metrics.compute()
metrics.reset()
m
```
