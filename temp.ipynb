```python
import tiktoken

# Load a pretrained BPE tokenizer
# Options include: "gpt2", "r50k_base", "p50k_base", "cl100k_base" (GPT-4/3.5), "o200k_base" (GPT-4o)
# enc = tiktoken.get_encoding("gpt2")

# Or load by model name (resolves to the right encoding automatically)
enc = tiktoken.encoding_for_model("gpt-4")

text = "This is some text"

# Encode to token IDs
ids = enc.encode(text)
print("IDs:", ids)
print("Count:", len(ids))

# Decode each token individually so you can see the segmentation
pieces = [enc.decode([i]) for i in ids]
print("Pieces:", pieces)

# Round-trip
print("Decoded:", enc.decode(ids))
```
